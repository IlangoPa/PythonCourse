```python
def UnivarFrequency(dataset,columnName):
    Total = len(dataset)
    freqTable = pd.DataFrame(columns=["Unique","Frequency","RelativeFreq%","Cumsum"])
    freqTable["Unique"]=dataset[columnName].value_counts().index
    freqTable["Frequency"] = dataset[columnName].value_counts().values
    freqTable["RelativeFreq%"] = freqTable["Frequency"]/Total*100
    freqTable["Cumsum"] = freqTable["RelativeFreq%"].cumsum()
    return freqTable
```

```python
UnivarFrequency(dataset,"ssc_p")
```
