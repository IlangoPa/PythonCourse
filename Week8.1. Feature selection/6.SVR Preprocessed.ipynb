```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset.columns
```

```python
independent = dataset.drop(['Profit'], axis =1)
```

```python
dependent = dataset['Profit']
```

```python
xtrain, xtest,ytrain,ytest = train_test_split(independent, dependent, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf', C= 1000)
regressor.fit(xtrain,ytrain)
```

```python
ypred = regressor.predict(xtest)
from sklearn.metrics import r2_score
rscore = r2_score(ytest,ypred)

rscore
```

```python
filename = "finalizedmodel.sav"
import pickle
pickle.dump(regressor,open(filename,'wb'))
```

```python
preinput = sc.transform([[1300,12000,4000,0,1]])
preinput
```
