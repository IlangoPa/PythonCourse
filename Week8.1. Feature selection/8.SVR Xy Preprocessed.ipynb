```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
dataset.columns
```

```python
independent = dataset.drop(['Profit'], axis =1)
```

```python
dependent = dataset['Profit']
```

```python
xtrain, xtest,ytrain,ytest = train_test_split(independent, dependent, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
scy = StandardScaler()
rs_ytrain = ytrain.values.reshape(-1,1)
rs_ytest = ytest.values.reshape(-1,1)
ytrain = scy.fit_transform(rs_ytrain)
ytest = scy.transform(rs_ytest)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf', C= 1000)
regressor.fit(xtrain,ytrain)
```

```python
ypred = regressor.predict(xtest)
from sklearn.metrics import r2_score
rscore = r2_score(ytest,ypred)

rscore
```

```python
filename = "finalizedmodel.sav"
import pickle
pickle.dump(regressor,open(filename,'wb'))

scfilename = 'scalerX.sav'
pickle.dump(sc,open(scfilename,'wb'))

scyfilename = 'scalery.sav'
pickle.dump(scy,open(scyfilename,'wb'))
```
